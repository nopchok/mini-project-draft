```python
all_df = {}
T123 = {}

result = None
for idx, folder in enumerate(list_folder):
    result = None
    start_row = list_start_row[idx]
    pattern_T0 = list( map(lambda x: (x[1], x[0][idx]), list_t0) )
    pattern_T1 = pattern_T0 + list( map(lambda x: (x[1], x[0][idx]), list_t1) )
    pattern_T2 = pattern_T0 + list( map(lambda x: (x[1], x[0][idx]), list_t2) )
    pattern_T3 = pattern_T0 + list( map(lambda x: (x[1], x[0][idx]), list_t3) )


    folder_path = '../data/' + str(folder)
    excel_files = [f for f in os.listdir(folder_path) if f.endswith(('.xlsx', '.xls'))]

    T123 = {}
    for file in excel_files:
        table = ''
        if 'T1' in file:
            table = 'T1'
            pattern = pattern_T1
        elif 'T2' in file:
            table = 'T2'
            pattern = pattern_T2
        elif 'T3' in file:
            table = 'T3'
            pattern = pattern_T3
        elif 'T4' in file:
            continue
        elif 'T5' in file:
            continue
        elif 'T6' in file:
            continue

        file_path = os.path.join(folder_path, file)

        df = pd.read_excel(file_path, header=None)
        df.columns = list(string.ascii_uppercase[:df.shape[1]])

        df = df[start_row-1:]

        patt = list( filter(lambda x: x[1] != None, pattern) )
        df = df[list(map(lambda x: x[1], patt))]
        df.columns = list(map(lambda x: x[0], patt))
        df['keep'] = np.select([df["สาขา"].isna()], [0], 1)
        df.ffill(inplace=True)
        df = df[df["keep"] == 1]
        df.drop(columns=['keep'], inplace=True)
        
        ex_cols = list(df.columns)
        df['ปี'] = folder
        df = df[['ปี'] + ex_cols]
        # print(df)
        
        patt_none = list( filter(lambda x: x[1] == None, pattern) )
        patt_none = list( map(lambda x: x[0], patt_none) )
        df = df.assign(**{col: None for col in patt_none})
        T123[table] = df

        if table == 'T3':
            result = T123['T1'].merge(
                T123['T2'],
                on=["ปี", "วิทยาเขต", "คณะ", "ภาค", "สาขา"],
                how="left"
            )
            result = result.merge(
                T123['T3'],
                on=["ปี", "วิทยาเขต", "คณะ", "ภาค", "สาขา"],
                how="left"
            )
            all_df[folder] = result
    # break
```

```python
# Concatenate vertically
all_result = pd.concat(all_df.values(), ignore_index=True)
```

```python
all_result.to_excel('all_result ป.ตรี.xlsx')
```
