```python
import pandas as pd
import numpy as np
```

```python
DF_1 = pd.read_excel('all_result ป.ตรี.xlsx', index_col=0)
DF_2 = pd.read_excel('all_result บัณฑิตศึกษา.xlsx', index_col=0)
```

```python
# checking column each files
tmp = pd.concat(
    [
        pd.DataFrame(True, columns=['file ป.ตรี'], index=DF_1.columns),
        pd.DataFrame(True, columns=['file บัณฑิตศึกษา'], index=DF_2.columns),
    ],
    axis=1
)

tmp[tmp.isna().any(axis=1)]
```

```python
# add value for ระดับ file ป.ตรี
tmp = DF_1.copy()
tmp['ระดับ'] = 'ป.ตรี'

df_concat = pd.concat([tmp, DF_2], axis=0)
print(df_concat.shape)
df_concat.columns
```

```python
df = df_concat.copy()

# columns integer to be clean
columns = ['ปี', 'ผู้สำเร็จการศึกษา',
'ผู้บันทึกข้อมูล', 'มีงานทำเดิม', 'ขอบเขตสำรวจ', 'ทำงาน', 'รัฐ',
'รัฐวิสาหกิจ', 'เอกชน', 'ธุรกิจส่วนตัว/อิสระ', 'ทำงานที่อื่นๆ',
'งานตรงสาขา', 'งานไม่ตรงสาขา', 'งานไม่ระบุสาขา', 'รายได้เฉลี่ย',
'องค์การระหว่างประเทศ', 'ไม่ทำงาน', 'รอคำตอบ', 'หางานไม่ได้',
'ศึกษาต่อ', 'ไม่ประสงค์ทำงาน', 'บวช', 'เกณฑ์ทหาร',
'ไม่ทำงาน สาเหตุอื่นๆ', 'ต้องการทำงานอิสระ']
for col in columns:
    df[col] = df[col].replace('-', '0')
    df[col] = df[col].fillna(0)
    df[col] = df[col].astype(int)
df.describe(include='all')
```

```python
df[['วิทยาเขต', 'ปี']].groupby('วิทยาเขต', dropna=False).count().sort_values('ปี')
```

```python
value_to_drop = [
    'ปริญญาโท',
    '5. โครงการจัดตั้งวิทยาเขตสุพรรณบุรี',
    '5.โครงการจัดตั้งวิทยาเขตสุพรรณบุรี',
    'สถาบันสมทบ',
    '5 สุพรรณบุรี',
]

df = df[~df['วิทยาเขต'].isin(value_to_drop)]
# original 4354
df.shape
```

```python
df['วิทยาเขต'].unique()
```

```python
# test
import re

tmp = df['วิทยาเขต'].unique().astype(str)
tmp = np.char.replace(tmp, ". ", ".")
tmp = np.char.replace(tmp, " ", ".")
tmp = np.array([re.sub(r'^[0-9]\.\s*', '', x) for x in tmp])
tmp = np.char.replace(tmp, ".", " ")
set(tmp)
```

```python
df['วิทยาเขต'] = (
    df['วิทยาเขต']
    .astype(str)
    .str.replace(". ", ".", regex=False)
    .str.replace(" ", ".", regex=False)
    .apply(lambda x: re.sub(r'^[0-9]\.\s*', '', x))
    .str.replace(".", " ", regex=False)
)
df['วิทยาเขต'].unique()
```

```python
df[['คณะ', 'ปี']].groupby('คณะ', dropna=False).count().sort_values('ปี')
```

```python
# test
import re

tmp = df['คณะ'].unique().astype(str)
tmp = np.char.replace(tmp, " กำแพงแสน", "")
tmp = np.char.replace(tmp, " ศรีราชา", "")
tmp = np.char.replace(tmp, "ศรีราชา", "")
tmp = np.char.replace(tmp, "และสุขภาพ", "")
set(tmp)
```

```python
df['คณะ'] = (
    df['คณะ']
    .astype(str)
    .str.replace(" กำแพงแสน", "")
    .str.replace(" ศรีราชา", "")
    .str.replace("ศรีราชา", "")
    .str.replace("และสุขภาพ", "")
)
df['คณะ'].unique()
```

```python
df[['ภาค', 'ปี']].groupby('ภาค', dropna=False).count().sort_values('ปี')
```

```python
df['ภาค'] = (
    df['ภาค']
    .str.replace(" ", "")
)
df['ภาค'].unique()
```

```python
df['ภาค'] = df['ภาค'].fillna('ภาคปกติ')
df[['ภาค', 'ปี']].groupby('ภาค', dropna=False).count().sort_values('ปี')
```

```python
df[['สาขา', 'ปี']].groupby('สาขา', dropna=False).count().sort_values('ปี')
```

```python
df['นานาชาติ'] = (
    df['สาขา'].astype(str).str.contains('นานาชาติ', na=False) |
    df['คณะ'].astype(str).str.contains('นานาชาติ', na=False)
)
df['นานาชาติ'].value_counts()
```

```python
set(df[df['นานาชาติ']]['สาขา'].unique())
```

```python
tmp = df[df['นานาชาติ']]['สาขา'].unique().astype(str)
tmp = np.char.replace(tmp, " (นานาชาติ)", " (หลักสูตรนานาชาติ)")
tmp = np.char.replace(tmp, " นานาชาติ", " (หลักสูตรนานาชาติ)")
tmp = np.char.replace(tmp, " หลักสูตรนานาชาติ", " (หลักสูตรนานาชาติ)")
tmp = np.char.replace(tmp, " (หลักสูตรนานาชาติ)", "")

mask = np.char.find(tmp, "นานา") != -1
not_contain = tmp[mask]
print(not_contain)
```

```python
df['สาขา'] = (
    df['สาขา']
    .astype(str).str.strip()
    .str.replace(" (นานาชาติ)", " (หลักสูตรนานาชาติ)")
    .str.replace(" นานาชาติ", " (หลักสูตรนานาชาติ)")
    .str.replace(" หลักสูตรนานาชาติ", " (หลักสูตรนานาชาติ)")
    .str.replace(" (หลักสูตรนานาชาติ)", "")
)
set(df['สาขา'].unique())
```

```python
set(df[df['สาขา'].str.contains(r"[^ ]\(.*\)", na=False)]['สาขา'].unique())
```

```python
df['สาขา'] = (
    df['สาขา']
    .astype(str)
    .str.replace("วิทยาศาสตรบัณฑิต(การจัดการเทคโนโลยีการบิน)", "วิทยาศาสตรบัณฑิต (การจัดการเทคโนโลยีการบิน)")
)
set(df[df['สาขา'].str.contains(r"[^ ]\(.*\)", na=False)]['สาขา'].unique())
```

```python
set(df[df['สาขา'].str.contains(r" \(.*\)", na=False)]['สาขา'].unique())
```

```python
tmp = df['สาขา'].astype(str).str.strip().unique().astype(str)
tmp = np.char.replace(tmp, ")", "")
tmp = np.char.replace(tmp, " (", " สาขา")
tmp = np.char.replace(tmp, "สาขาวิชา", "สาขา")

mask = np.char.find(tmp, "สาขาวิชา") != -1
not_contain = tmp[mask]
set(not_contain)
```

```python
df['สาขา'] = (
    df['สาขา']
    .astype(str)
    .str.strip()
    .str.replace(")", "")
    .str.replace(" (", " สาขา")
    .str.replace("สาขาวิชา", "สาขา")
    .str.replace("  ", " ")
)
set(df['สาขา'].unique())
```
