```python
import pandas as pd
import numpy as np
import os
import string
```

```python
list_folder = [57, 58, 59, 60, 61,     62, 63, 64, 65, 66]
# start_row
list_start_row = [9, 9, 9, 10, 8,    7, 7, 7, 7, 7]
list_t0 = []
list_t0.append((['A', 'C', 'C', 'B', 'C',    'B', 'B', 'D', 'D', 'C'], 'ระดับ'))
list_t0.append((['A', 'D', 'D', 'C', 'C',    'C', 'C', 'E', 'E', 'D'], 'วิทยาเขต'))
list_t0.append((['B', 'E', 'E', 'D', 'D',    'D', 'D', 'F', 'F', 'E'], 'คณะ'))
list_t0.append((['C', 'F', 'F', None, None, None, None, None, None, None], 'ภาค'))
list_t0.append((['D', 'G', 'G', 'E', 'E',    'E', 'E', 'G', 'G', 'F'], 'สาขา'))

list_t1 = []
list_t1.append((['E', 'H', 'H', 'F', 'F',    'F', 'F', 'H', 'H', 'G'], 'ผู้สำเร็จการศึกษา'))
list_t1.append((['F', 'I', 'I', 'G', 'G',    'G', 'G', 'I', 'I', 'H'], 'ผู้บันทึกข้อมูล'))
list_t1.append((['H', 'K', 'K', 'I', 'I',    'I', 'I', 'L', 'K', 'J'], 'มีงานทำเดิม'))
list_t1.append((['M', 'P', 'P', 'N', 'N',    'N', 'N', 'P', 'O', 'N'], 'ขอบเขตสำรวจ'))

list_t2 = []
list_t2.append((['E', 'H', 'H', 'F', 'F',    'F', 'F', 'H', 'H', 'G'], 'ทำงาน'))
list_t2.append((['F', 'I', 'I', 'G', 'G',    'G', 'G', 'I', 'I', 'H'], 'รัฐ'))
list_t2.append((['H', 'K', 'K', 'I', 'I',    'I', 'I', 'K', 'K', 'J'], 'รัฐวิสาหกิจ'))
list_t2.append((['J', 'M', 'M', 'K', 'K',    'K', 'K', 'M', 'M', 'L'], 'เอกชน'))
list_t2.append((['L', 'O', 'O', 'M', 'M',    'M', 'M', 'O', 'O', 'N'], 'ธุรกิจส่วนตัว/อิสระ'))
list_t2.append(([None, 'Q', 'Q', 'O', 'O',   'O', 'O', None, None, 'P'], 'องค์การระหว่างประเทศ'))
list_t2.append((['N', 'S', 'S', 'Q', None,   'Q', None, 'Q', 'Q', 'R'], 'ทำงานที่อื่นๆ'))
list_t2.append((['P', 'U', 'U', 'S', 'Q',    'S', 'Q', 'S', 'S', 'T'], 'งานตรงสาขา'))
list_t2.append((['R', None, None, None, 'S',  'U', None, 'U', 'U', 'V'], 'งานไม่ตรงสาขา'))
list_t2.append((['T', None, None, None, 'U',  None, None, None, None, None], 'งานไม่ระบุสาขา'))
list_t2.append((['V', 'W', 'W', 'U', 'W',    'W', 'S', 'W', 'W', 'X'], 'รายได้เฉลี่ย'))

list_t3 = []
list_t3.append((['E', 'H', 'H', 'F', 'F',    'F', 'F', 'H', 'H', 'G'], 'ไม่ทำงาน'))
list_t3.append((['F', 'Q', 'K', 'O', 'K',    'O', 'I', 'K', 'K', 'J'], 'รอคำตอบ'))
list_t3.append((['H', 'S', 'M', 'Q', 'M',    'Q', 'K', 'M', 'M', 'L'], 'หางานไม่ได้'))
list_t3.append((['J', 'I', 'S', 'G', 'G',    'G', 'S', 'U', 'U', 'T'], 'ศึกษาต่อ'))
list_t3.append((['L', 'K', 'I', 'I', 'I',    'I', 'G', 'I', 'I', 'H'], 'ไม่ประสงค์ทำงาน'))
list_t3.append((['N', 'M', 'O', 'K', 'Q',    'K', 'O', 'Q', 'Q', 'P'], 'บวช'))
list_t3.append((['P', 'O', 'Q', 'M', 'S',    'M', 'Q', 'S', 'S', 'R'], 'เกณฑ์ทหาร'))
list_t3.append(([None, None, None, None, 'O','S', 'M', 'O', 'O', 'N'], 'ต้องการทำงานอิสระ'))
list_t3.append((['R', 'U', 'U', 'U', 'U',    'U', 'U', 'W', 'W', 'V'], 'ไม่ทำงาน สาเหตุอื่นๆ'))

# fix table '64
for list_t in [list_t0, list_t1, list_t2, list_t3]:
    for i in range(len(list_t)):
        x = list_t[i][0][7]
        if x is not None:
            list_t[i][0][7] = chr(ord(x) - 1)
```

```python
all_df = {}
T123 = {}

result = None
for idx, folder in enumerate(list_folder):
    result = None
    start_row = list_start_row[idx]
    pattern_T0 = list( map(lambda x: (x[1], x[0][idx]), list_t0) )
    pattern_T1 = pattern_T0 + list( map(lambda x: (x[1], x[0][idx]), list_t1) )
    pattern_T2 = pattern_T0 + list( map(lambda x: (x[1], x[0][idx]), list_t2) )
    pattern_T3 = pattern_T0 + list( map(lambda x: (x[1], x[0][idx]), list_t3) )

    folder_path = '../data/' + str(folder)
    excel_files = [f for f in os.listdir(folder_path) if f.endswith(('.xlsx', '.xls'))]

    T123 = {}
    for file in excel_files:
        table = ''
        if 'T4' in file:
            table = 'T4'
            pattern = pattern_T1
        elif 'T5' in file:
            table = 'T5'
            pattern = pattern_T2
        elif 'T6' in file:
            table = 'T6'
            pattern = pattern_T3
        else:
            continue

        file_path = os.path.join(folder_path, file)

        df = pd.read_excel(file_path, header=None)
        df.columns = list(string.ascii_uppercase[:df.shape[1]])

        df = df[start_row-1:]

        patt = list( filter(lambda x: x[1] != None, pattern) )
        df = df[list(map(lambda x: x[1], patt))]
        df.columns = list(map(lambda x: x[0], patt))
        
        # fix ปริญญา column
        if folder in (57, 61):
            df.loc[~df["ระดับ"].str.contains("ปริญญา", na=False) & ~df["ระดับ"].str.contains("บัณฑิต", na=False), "ระดับ"] = np.nan
            # print(df)
            # break
        
        df['keep'] = np.select([df["สาขา"].isna()], [0], 1)
        df.ffill(inplace=True)
        df = df[df["keep"] == 1]
        df.drop(columns=['keep'], inplace=True)
        
        ex_cols = list(df.columns)
        df['ปี'] = folder
        df = df[['ปี'] + ex_cols]
        # print(df)
        
        patt_none = list( filter(lambda x: x[1] == None, pattern) )
        patt_none = list( map(lambda x: x[0], patt_none) )
        df = df.assign(**{col: None for col in patt_none})
        T123[table] = df

        if table == 'T6':
            result = T123['T4'].merge(
                T123['T5'],
                on=["ปี", "ระดับ", "วิทยาเขต", "คณะ", "ภาค", "สาขา"],
                how="left"
            )
            result = result.merge(
                T123['T6'],
                on=["ปี", "ระดับ", "วิทยาเขต", "คณะ", "ภาค", "สาขา"],
                how="left"
            )
            all_df[folder] = result
    # break
```

```python
# Concatenate vertically
all_result = pd.concat(all_df.values(), ignore_index=True)
```
